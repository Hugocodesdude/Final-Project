# Data Extraction 

```python
import os
import requests
from bs4 import BeautifulSoup as bs4
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from bs4 import BeautifulSoup as bs4
```

```python
def list_of_links (link):
    res = requests.get(link)
    soup = bs4(res.content, 'html.parser')
    return [i.find_all('a')[0].get('href') for i in soup.find_all('dl')[0].find_all('dt')]

def save_into_dir (dir_name, list_):
    for i in list_:
        os.system(f'curl {i} >> {dir_name}/{i.split("/")[-1]}')
```

# Amsterdam

- CO, PM10, PM2, NO2

## CO

```python
amsterdam_co = list_of_links ('https://fme.discomap.eea.europa.eu/fmedatastreaming/AirQualityDownload/AQData_Extract.fmw?CountryCode=NL&CityName=Amsterdam&Pollutant=10&Year_from=2013&Year_to=2022&Station=&Samplingpoint=&Source=E1a&Output=HTML&UpdateDate=&TimeCoverage=Year')
```

```python
save_into_dir ('amsterdam_co', amsterdam_co)
```
